Converting the RGB image to GrayScale image using openCV


```python
#importing opencv library
import cv2
```

```python
img=cv2.imread('/content/strangerThings_image_resized.jpg')
```

```python
img.shape
```

```python
grayscale_image=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
grayscale_image.shape
```

```python
cv2.imwrite("strangerThings_grey.jpg", grayscale_image) #for saving the image
```
