## Linear Regression,Ridge And Lasso

```python
#housing Princing 
from sklearn.datasets import fetch_california_housing
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=fetch_california_housing()
```

```python
dataset=pd.DataFrame(df.data)
dataset.columns=df.feature_names
dataset.head()
```

```python
df.keys()
```

### Here all are independent feature except Price(It is dependent feature)

```python
dataset['Price']=df.target
dataset.head()
```

```python
##Dividing the dataset into independent and dependent feature
X=dataset.iloc[:,:-1] #independent features
Y=dataset.iloc[:,-1] #dependent feature
print(X)
print(Y)
```

```python
##Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_reg=LinearRegression()
mse=cross_val_score(lin_reg,X,Y,scoring='neg_mean_squared_error',cv=5) #split into 5 folds
mean_mse=np.mean(mse)
print(mean_mse)
```

```python
#Ridge Reggression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge=Ridge()
params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20]}
ridge_regressor=GridSearchCV(ridge,params,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(X,Y)



```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

```python
#Lasso Regression
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso=Lasso()
params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20]}
lasso_regressor=GridSearchCV(lasso,params,scoring='neg_mean_squared_error',cv=5)
lasso_regressor.fit(X,Y)

```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.33,random_state=42)
```

```python
##Linear Regreesion
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_reg=LinearRegression()
mse=cross_val_score(lin_reg,X_train,Y_train,scoring='neg_mean_squared_error',cv=5) #split into 5 folds
mean_mse=np.mean(mse)
print(mean_mse)
```

```python
#Ridge Reggression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge=Ridge()
params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20]}
ridge_regressor=GridSearchCV(ridge,params,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(X_train,Y_train)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

```python
#Lasso Regression
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso=Lasso()
params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20]}
lasso_regressor=GridSearchCV(lasso,params,scoring='neg_mean_squared_error',cv=5)
lasso_regressor.fit(X_train,Y_train)
```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
y_pred=lasso_regressor.predict(X_test)
from sklearn.metrics import r2_score
r2_score1 = r2_score(Y_test, y_pred)
```

```python
print(r2_score1)
```

```python
y_pred=ridge_regressor.predict(X_test)
from sklearn.metrics import r2_score
r2_score1 = r2_score(Y_test, y_pred)
print(r2_score1)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import GridSearchCV
```

```python
df=load_breast_cancer()
### Independent feature
X=pd.DataFrame(df.data,columns=df.feature_names)
```

```python
X.head()
```

```python
##Dependent feature
Y=pd.DataFrame(df['target'],columns=["Target"])
```

```python
Y.head()
```

```python
Y['Target'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.33,random_state=42)
```

```python
params=[{'C':[1,5,10]},{'max_iter':[100,150]}]
```

```python
model1=LogisticRegression(C=100,max_iter=100)
```

```python
model=GridSearchCV(model1,param_grid=params,scoring='f1',cv=5)
```

```python
model.fit(X_train,Y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
confusion_matrix(Y_test,y_pred)
```

```python
accuracy_score(Y_test,y_pred)
```

```python
print(classification_report(Y_test,y_pred))
```
